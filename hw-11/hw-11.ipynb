### Задание 1
#### Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
# подгружаем данные по рейтингам
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings.info()
```

```python
# функция возвращает текст в зависимости от входящей оценки
def class_fnc(param):
    if param <= 2:
        return 'низкий рейтинг'
    elif 2 < param <= 4: 
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
    
    return 'что-то пошло не так'
```

```python
# создаем новый столбец используя функцию class_fnc
ratings['class'] = ratings['rating'].apply(class_fnc)
ratings.head(10)
```

```python
# выводим результат
ratings.groupby('class').count().reset_index()[['class', 'rating']]
```

### Задание 2
Используем файл keywords.csv.

#### Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

 geo_data = {
 'Центр': ['москва', 'тула', 'ярославль'],
 'Северо-Запад': ['петербург', 'псков', 'мурманск'],
 'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
 }

Результат классификации запишите в отдельный столбец region.

```python
# подгружаем данные по рейтингам
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
keywords.info()
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] 
}
```

```python
a = '4Fdsks9 sfdlkjfsd dsfldsjfdslf'
```

```python
for i in a.split():
    print(i)

a.split()
```

```python
# функция проходит до всем записям словаря и проверят входящий параметр
# на вхождение в "значение", если находит возвращает "ключ", если нет - 'undefined'
# при этом входящий параметр разбивается на слова и проверяется каждое слово отдельно
# дополнительно - сделал регистронезависимым (хотя городов с Большой буквы в запросах не нашлось)
def geo_class(param):
    for i in param.split():
        for key, value in geo_data.items():
            if i.lower() in value: 
                return key
    
    return 'undefined'
```

```python
# создаем новый столбец используя функцию geo_class
keywords['region'] = keywords['keyword'].apply(geo_class)
keywords.head(10)
```

```python
# выводим результат
keywords.groupby('region').count().reset_index()
```

```python
# еще небольшой вывод данных
keywords[keywords['region'] != 'undefined']
```
