#### Задание 2. Работа с выбросами
> В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
# rectal temperature
q1 = df_short['rectal temperature'].quantile(0.25)
q3 = df_short['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_short[df_short['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal temperature').describe()
remove_outliers
```

```python
# выбросами по rectal temperature признаны 14 записей. видно, что были выброшены и верхние и нижние значения. медиана и средняя не изменились. думаю, что их можно оставить.
```

```python
# pulse
q1 = df_short['pulse'].quantile(0.25)
q3 = df_short['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_short[df_short['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse').describe()
remove_outliers
```

```python
# выбросами по pulse признаны 5 записей. видно, что были выброшены только верхние значения. медиана и средняя изменились незначительно. учитывая, что качественно ничего не меняется, считаю данные надо оставить в полном объеме
```

```python
# respiratory rate
q1 = df_short['respiratory rate'].quantile(0.25)
q3 = df_short['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_short[df_short['respiratory rate'].between(lower_bound, upper_bound, inclusive=True)].sort_values('respiratory rate').describe()
remove_outliers
```

```python
# выбросами по respiratory rate признаны 17 записей. видно, что были выброшены только верхние значения. медиана изменилась незначительно, средняя чуть более значительно.
# отброшенные данные в данном случае экстремальны, но медиану практически не повлияли. 
# с одной стороны их можно не учитывать, с другой стороны "природа" их экстремальности не очевидна при этом на значимые для меня показатели (медиана) они значимо не влияют
# нет у меня пока четкого понимания что с ними делать, поэтому лучше пока оставить
```
