### Задание 1

> ##### Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:
если источник traffic_source равен yandex или google, то в source_type ставится organic\
для источников paid и email из России - ставим ad\
для источников paid и email не из России - ставим other\
все остальные варианты берем из traffic_source без изменений

```python
# подгружаем данные по логов
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
log.info()
```

```python
organic = ['yandex', 'google']
country = ['Russia']
other = ['paid', 'email']
```

```python
# функция определяет источник траффика
def source_type_fnc(row):
    if row.traffic_source in organic: # проверяем на яндекс и гугл
        return 'organic'
    else:
        if row.traffic_source in other:  # проверяем на оплаченный и е-рассылки
            if row.region in country: # проверяем на страну
                return 'ad'
            else:
                return 'other'
        else:
            return row.traffic_source
    
    return 'что-то пошло не так'
```

```python
# создаем новый столбец используя функцию source_type_fnc
log['source_type'] = log.apply(source_type_fnc, axis=1)
log.head(10)
```

```python
# выводим результат
log.groupby('source_type').count().reset_index()
```

### Задание 3

> ##### Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. \
Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import numpy as np
```

```python
# подгружаем данные по рейтингам
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings.info()
```

### возможно моё решение "костыльное", если да - буду признателен за подсказки "хороших" решений

```python
# объявляем лимит количества оценок для расчета среднего времени жизни
limit = 100
# объявляем новый словарь
timelife = {}
# создаем список уникальных пользователей
unique_users = set(ratings.userId)
```

```python
# проходимся по каждому уникальному пользователю из списка
for user in unique_users:
    if len(ratings.loc[ratings['userId']== user]) > limit: # если пользователь проставил оценок больше чем наш лимит
        minim = ratings.loc[ratings['userId']== user, 'timestamp'].min() # сохраняем время его первой оценки
        maxim = ratings.loc[ratings['userId']== user, 'timestamp'].max() # сохраняем время его крайней оценки
        timelife[user] = maxim-minim # добавляем в словарь пару пользователь: время жизни
```

```python
# смотрим сколько записей в словаре
len(timelife)
```

```python
# считаем среднее время жизни по пользователям
np.array(list(timelife.values())).mean()
```

> ##### Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:
таблицу с тремя типами выручки для каждого client_id без указания адреса клиента\
аналогичную таблицу по типам выручки с указанием адреса клиента\
Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# объединяем данные через merge. выбрал outer чтобы ничего не потерялось
union_1 = rzd.merge(auto, how = 'outer', left_on='client_id', right_on='client_id').merge(air, how = 'outer', left_on='client_id', right_on='client_id')
union_1.head(10)
```

```python
# объединяем данные через merge. выбрал outer чтобы ничего не потерялось (начинаем с таблицы адресов клиентов)
union_2 = client_base.merge(rzd, how = 'outer', left_on='client_id', right_on='client_id').merge(auto, how = 'outer', left_on='client_id', right_on='client_id').merge(air, how = 'outer', left_on='client_id', right_on='client_id')
union_2.head(10)
```
